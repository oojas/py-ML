# Dropping Constant values

```python
import pandas as pd
```

```python
data=pd.read_csv("Santander.csv",nrows=10000)
```

```python
data
```

```python
[col for col in data.columns if data[col].isnull().sum() > 0]
```

```python
data.shape
```

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
var_thres=VarianceThreshold(threshold=0)
```

```python
var_thres.fit(data)
```

```python
var_thres.get_support()
```

```python
constant_val=[col for col in data.columns
              if col not in data.columns[var_thres.get_support()]]
```

```python
constant_val
```

```python
data.drop(constant_val,axis=1,inplace=True)
```
