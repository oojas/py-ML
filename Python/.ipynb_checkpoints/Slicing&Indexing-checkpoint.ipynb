# Slicing and Indexing

```python
import pandas as pd
```

```python
from sklearn.datasets import load_boston
```

```python
boston=load_boston()
```

```python
df=pd.DataFrame(boston.data,columns=[boston.feature_names])
```

```python
df.head()
```

```python
df[1:3]
```

```python
df[3:]
```

```python
df[:6]
```

```python
df.loc[:2]
```

```python
df.loc[:2,['CRIM','ZN']]
```
